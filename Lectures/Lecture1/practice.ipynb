# Activities

Note: the practice happens almost exclusively on paper, potentially augmented by multiple choice questions in order to submit an answer that shows understanding.

The activities performed are:
- Translate the following states into natural language
    - $\{ position1:= \{ x := 3, y := 0 \}, position2:= \{ x := 5, y := \text{"De Jong"} \} \}$
    - $\{ person1:= \{ name := \text{"Jim"}, surname := \text{"Smit"} \}, person2:= \{ name := \text{"Wim"}, surname := \text{"De Jong"} \} \}$
    - $\{ mondayForecast:= \{ condition := \text{"Cloudy"}, minTemperature := 13, maxTemperature := 23, wind := 15, humidity := 63\% \},
     sundayForecast:= \{ condition := \text{"Sunny"}, minTemperature := 18, maxTemperature := 26, wind := 0, humidity := 80\% \}\}$
    - $\{ student1:= \{ name := \text{"John"}, surname := \text{"Doe"}, course := \{ code := \text{"CMIINF"}, name := \text{"Computer Science"} \}  \}, 
    student2:= \{ name := \text{"Jane"}, surname := \text{"Doe"}, course := \{ code := \text{"CMITI"}, name := \text{"Computer Engineering"} \}  \} \}$    
- Modify the state representation of the `Turtle` program to include the rotation of the turtle in the state.
- Model the following real world information
   - A car has a model name, a specification saying whether it is used or not, and the mileage (total distance travelled during its lifetime).
   - A video game has a title, a producer, a PEGI (https://en.wikipedia.org/wiki/Pan_European_Game_Information), and a price.
   - A CPU has a model code, an amout of cache levels, an amount of cores, and the frequency. (example: https://ark.intel.com/products/97185/Intel-Core-i7-7700HQ-Processor-6M-Cache-up-to-3_80-GHz)
   - Modify the solution of the exercise above to explicitly represent the level of the cache and the size of each cache level (again refer to the link above).
   - Represent a citizen with BSN, name, surname, address, and job. An address contains the city, street, and house number. The job contains a name, a salary, and a scale level (which is a number).

*For the following exercises do not use GrandeOmega. Assume that, by convention, the horizontal axis points to the right and the vertical axis points to the top. Also remember that in "Turtle" programming language the state contains a variable p that stores the state of the pen.*

1. Given the state $\{ x := 5, y := -2, p := \text{"on"} \}$, where x and y represent the position of a student, write the state trace (the representation of the state after executing each statement) of the following program and draw the picture rendered by the pen in the reference system.
    ```
    up 4;
    right 1;
    down 1;
    done;
    ```
